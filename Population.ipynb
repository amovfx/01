```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
```

```python
df = pd.read_csv('population.csv')
```

```python
df.head()
```

```python
f_df = df[df['TIME_PERIOD'] == 2021]
f_df.head()
```

```python
age_df = f_df[f_df['AGE'].str.contains('^Y\d+.*T\d+$', regex=True)]
age_df
```

```python
babies = f_df.loc[f_df['AGE'] == 'Y_LE4'].reset_index(drop=True)
oldies = f_df.loc[f_df['AGE'] == 'Y_GE85'].reset_index(drop=True)
```

```python
babies.loc[0, 'AGE'] = 'Y0T4'
oldies.loc[0, 'AGE'] = 'Y85T125'
oldies
```

```python
cleaned_ages = pd.concat([age_df, babies, oldies], ignore_index=True).reset_index(drop=True)
cleaned_ages
```

```python
cleaned_ages['start_age'] = cleaned_ages['AGE'].str.extract('Y(\d+)T', expand=False).astype(int)
cleaned_ages['end_age'] = cleaned_ages['AGE'].str.extract('T(\d+)', expand=False).astype(int)

cleaned_ages
```

```python
cleaned_ages = cleaned_ages.sort_values('start_age').reset_index(drop=True)
cleaned_ages
```

```python
plt.plot(cleaned_ages['sorter'], cleaned_ages['OBS_VALUE'])
plt.show()
```

```python
cleaned_ages['difference'] = cleaned_ages['end_age'] - cleaned_ages['start_age']
cleaned_ages 
```

```python
cleaned_ages = cleaned_ages[cleaned_ages['difference'] < 41].reset_index(drop=True)
```
