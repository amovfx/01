```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
```

# Import population data

```python
df = pd.read_csv('population.csv')
```

```python
df.head()
```

# Filter the year 2021

```python
f_df = df[df['TIME_PERIOD'] == 2021]
f_df.head()
```

# Filter all age ranges that dont match the pattern Y{XX}D{XX}

```python
age_df = f_df[f_df['AGE'].str.contains('^Y\d+.*T\d+$', regex=True)]
age_df
```

# Keep the babies and old people

```python
babies = f_df.loc[f_df['AGE'] == 'Y_LE4'].reset_index(drop=True)
oldies = f_df.loc[f_df['AGE'] == 'Y_GE85'].reset_index(drop=True)
```

### Change the age data to the convention of Y{XX}D{XX}

```python
babies.loc[0, 'AGE'] = 'Y0T4'
oldies.loc[0, 'AGE'] = 'Y85T125'
oldies
```

# Expand the the Y85-Y125 into an exponential distribution

```python
from scipy.stats import expon

rv = expon(loc=0, scale=5)
x = np.linspace(0, 25, 100)
pdf = rv.pdf(x)
plt.plot(x, pdf)
plt.show()
```

```python
def interval(start):
    return f'Y{85+start}T{85+start+4}', int((rv.cdf(start + 5) - rv.cdf(start)) * oldies['OBS_VALUE'].sum())

vals = [interval(i)  for i in range(0,25,5)]
df_extrapolated_geriatrics = pd.DataFrame(vals, columns=['AGE', 'OBS_VALUE'])
```

```python
cleaned_ages = pd.concat([babies, age_df, df_extrapolated_geriatrics], ignore_index=True).reset_index(drop=True)
df_ages = pd.concat([cleaned_ages['AGE'], cleaned_ages['OBS_VALUE']], axis=1)
df_ages
```

```python
df_ages['start_age'] = df_ages['AGE'].str.extract('Y(\d+)T', expand=False).astype(int)
df_ages['end_age'] = df_ages['AGE'].str.extract('T(\d+)', expand=False).astype(int)

df_ages
```

# Remove rows where the difference of age is greater than 5 and sort

```python
condition = df_ages['end_age'] - df_ages['start_age'] < 5
df_age_brackets = df_ages[condition].reset_index(drop=True)
df_age_brackets = df_age_brackets.drop(['end_age'], axis=1)
df_age_brackets = df_age_brackets.sort_values(by=['start_age']).reset_index(drop=True)
df_age_brackets
```

# Graph the population distribution

```python
plt.plot(df_age_brackets['start_age'], df_age_brackets['OBS_VALUE'])
plt.show()
```
